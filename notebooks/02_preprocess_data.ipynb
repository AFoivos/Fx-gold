```python
import pandas as pd
import matplotlib.pyplot as plt
import os
print(os.getcwd())

df_1h = pd.read_csv('../data/XAU_1h_data.csv', sep=';')
df_1d = pd.read_csv('../data/XAU_1d_data.csv', sep=';')

df_1h['Date'] = pd.to_datetime(df_1h['Date'], format="%Y.%m.%d %H:%M")
df_1d['Date'] = pd.to_datetime(df_1d['Date'], format="%Y.%m.%d %H:%M")

df_1h.set_index('Date', inplace=True)
df_1d.set_index('Date', inplace=True)

df_1h.sort_index(inplace=True)
df_1d.sort_index(inplace=True)

print("Null values in 1h data:")
print(df_1h.isnull().sum())
print("\nNull values in 1d data:")
print(df_1d.isnull().sum())

df_1h['Return'] = df_1h['Close'].pct_change()
df_1d['Return'] = df_1d['Close'].pct_change()

df_1h['Return'].plot(figsize=(12,4), title='1H Returns')
plt.show()

df_1d['Return'].plot(figsize=(12,4), title='1D Returns')
plt.show()

df_1h.to_csv('../data/XAU_1h_clean.csv')
df_1d.to_csv('../data/XAU_1d_clean.csv')
```
